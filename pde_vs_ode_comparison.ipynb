# PDE vs ODE Approximation Comparison

## Purpose

Compare the full HJB PDE solution against the ODE-based quadratic approximation from [1] to evaluate the accuracy of the approximation and validate the PDE solver.

## Planned Sections

1. **Setup** - Load/run both the PDE solver and the ODE approximation for matching parameters (2-currency case).
2. **Value function comparison** - Compare the PDE value function against the quadratic approximation $\hat{\theta}(t,y) = -y^T A(t) y - y^T B(t) - C(t)$.
3. **Optimal quotes comparison** - Compare optimal client markups from PDE vs ODE across inventory levels.
4. **Hedging rate comparison** - Compare optimal hedging rates from PDE vs ODE.
5. **Error analysis** - Quantify approximation error (pointwise, L2, relative) as a function of inventory size.
6. **Regime exploration** - Identify parameter regimes where the ODE approximation breaks down (e.g., large inventory, high volatility, strong cross-currency correlation).
7. **Monte Carlo P&L comparison** - Simulate inventory paths under both policies and compare realized P&L distributions.
8. **Summary and conclusions** - Characterize when the ODE approximation is sufficient vs when the full PDE is needed.
