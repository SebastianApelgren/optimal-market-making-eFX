# Physics-Informed Neural Networks (PINNs) for the HJB Equation

## Purpose

Solve the HJB PDE using physics-informed neural networks to handle higher-dimensional settings (beyond 2-3 currencies) without grid-based discretization. Compare PINN solutions against the grid-based PDE solver and ODE approximation.

## Planned Sections

1. **PINN formulation** - Encode the HJB PDE as a loss function for the neural network.
2. **Network architecture** - Design the neural network (inputs: time + inventory, output: value function).
3. **Training setup** - Collocation points, boundary/terminal conditions, optimizer, learning rate schedule.
4. **2-currency benchmark** - Train on the 2-currency case and compare against the grid-based PDE solution.
5. **Higher-dimensional experiments** - Scale to 3-5 currencies and assess accuracy.
6. **Extract optimal controls** - Compute quotes and hedging rates from the trained PINN via automatic differentiation.
7. **Comparison with ODE approximation** - How does the PINN solution compare to the closed-form ODE result?
8. **Computational cost analysis** - Training time, inference time, and accuracy trade-offs vs grid-based methods.
9. **Summary and conclusions** - Assess viability of PINNs for this class of problems.
